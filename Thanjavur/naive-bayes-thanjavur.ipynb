# WITHOUT FEATURE SELECTION

```python
# âœ… Import Required Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, QuantileTransformer
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score, 
    confusion_matrix, roc_curve, auc
)
from imblearn.metrics import geometric_mean_score

# âœ… Load dataset
df = pd.read_csv("/kaggle/input/thanjavur-dataset-new/updated_dataset .csv")  # Change to your dataset

# âœ… Remove Missing Values
df.dropna(inplace=True)

# âœ… Remove Outliers using IQR
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
df = df[~((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).any(axis=1)]

# âœ… Define Features & Target
# ðŸ”¥ Removing highly correlated features (`T2MDEW`, `RainToday`)
X = df.drop(columns=["RainTomorrow", "T2MDEW", "RainToday"])
y = df["RainTomorrow"]

# âœ… Normalize Data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# âœ… Apply Quantile Transformation
qt = QuantileTransformer(output_distribution="normal", random_state=42)
X_transformed = qt.fit_transform(X_scaled)

# âœ… Apply PCA (Keep 95% Variance)
pca = PCA(n_components=0.93, random_state=42)
X_pca = pca.fit_transform(X_transformed)
print(f"\nðŸ”¹ PCA Components Retained: {pca.n_components_}")

# âœ… Apply Clustering (KMeans)
n_clusters = 3  # You can tune this
kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
cluster_labels = kmeans.fit_predict(X_pca)

# âœ… Add Cluster Labels to Data
X_final = np.hstack((X_pca, cluster_labels.reshape(-1, 1)))

# âœ… Train-Test Split (Stratified)
X_train, X_test, y_train, y_test = train_test_split(
    X_final, y, test_size=0.2, random_state=42, stratify=y
)

# âœ… Train NaÃ¯ve Bayes Model
nb_model = GaussianNB()
nb_model.fit(X_train, y_train)

# âœ… Predictions
y_pred = nb_model.predict(X_test)
y_pred_prob = nb_model.predict_proba(X_test)[:, 1]

# âœ… Compute Metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
g_mean = geometric_mean_score(y_test, y_pred)

# âœ… Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(5, 4))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=["No Rain", "Rain"], yticklabels=["No Rain", "Rain"])
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()

# âœ… ROC Curve
fpr, tpr, _ = roc_curve(y_test, y_pred_prob)
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(6, 5))
plt.plot(fpr, tpr, color="blue", lw=2, label=f"ROC curve (AUC = {roc_auc:.2f})")
plt.plot([0, 1], [0, 1], color="gray", linestyle="--")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("Receiver Operating Characteristic (ROC)")
plt.legend(loc="lower right")
plt.show()

# âœ… Print Performance Metrics
print("\nðŸ“Š Model Performance Metrics:")
print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1-Score: {f1:.4f}")
print(f"G-Mean: {g_mean:.4f}")
```

# WITH FEATURE SELECTION

```python
# âœ… Import Required Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, QuantileTransformer
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.naive_bayes import GaussianNB
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score, 
    confusion_matrix, roc_curve, auc
)
from imblearn.metrics import geometric_mean_score

# âœ… Load dataset
df = pd.read_csv("/kaggle/input/thanjavur-dataset-new/updated_dataset .csv")  # Change to your dataset

# âœ… Remove Missing Values
df.dropna(inplace=True)

# âœ… Remove Outliers using IQR
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
df = df[~((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).any(axis=1)]

# âœ… Define Features & Target
# ðŸ”¥ Removing highly correlated features (`T2MDEW`, `RainToday`)
X = df.drop(columns=["RainTomorrow", "T2MDEW", "RainToday","MO"])
y = df["RainTomorrow"]

# âœ… Normalize Data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# âœ… Feature Selection: Select Top K Features
k = 4 # Choose top 4 best features
selector = SelectKBest(score_func=f_classif, k=k)
X_selected = selector.fit_transform(X_scaled, y)
selected_features = X.columns[selector.get_support()]
print("\nðŸ”¥ Selected Features:", list(selected_features))

# âœ… Apply Quantile Transformation
qt = QuantileTransformer(output_distribution="normal", random_state=42)
X_transformed = qt.fit_transform(X_selected)

# âœ… Apply PCA (Keep 95% Variance)
pca = PCA(n_components=0.95, random_state=42)
X_pca = pca.fit_transform(X_transformed)
print(f"\nðŸ”¹ PCA Components Retained: {pca.n_components_}")

# âœ… Apply Clustering (KMeans)
n_clusters = 3  # You can tune this
kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
cluster_labels = kmeans.fit_predict(X_pca)

# âœ… Add Cluster Labels to Data
X_final = np.hstack((X_pca, cluster_labels.reshape(-1, 1)))

# âœ… Train-Test Split (Stratified)
X_train, X_test, y_train, y_test = train_test_split(
    X_final, y, test_size=0.2, random_state=42, stratify=y
)

# âœ… Train NaÃ¯ve Bayes Model
nb_model = GaussianNB()
nb_model.fit(X_train, y_train)

# âœ… Predictions
y_pred = nb_model.predict(X_test)
y_pred_prob = nb_model.predict_proba(X_test)[:, 1]

# âœ… Compute Metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
g_mean = geometric_mean_score(y_test, y_pred)

# âœ… Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(5, 4))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=["No Rain", "Rain"], yticklabels=["No Rain", "Rain"])
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()

# âœ… ROC Curve
fpr, tpr, _ = roc_curve(y_test, y_pred_prob)
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(6, 5))
plt.plot(fpr, tpr, color="blue", lw=2, label=f"ROC curve (AUC = {roc_auc:.2f})")
plt.plot([0, 1], [0, 1], color="gray", linestyle="--")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("Receiver Operating Characteristic (ROC)")
plt.legend(loc="lower right")
plt.show()

# âœ… Print Performance Metrics
print("\nðŸ“Š Model Performance Metrics:")
print(f"Accuracy After Feature Selection: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1-Score: {f1:.4f}")
print(f"G-Mean: {g_mean:.4f}")
```
